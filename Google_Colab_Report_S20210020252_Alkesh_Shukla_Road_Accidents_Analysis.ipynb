**Code Explanation**


1. **Selecting Numerical Columns:**
   - We pick only the columns with numbers from our dataset to focus on variables that can be correlated.

2. **Computing Correlation Matrix:**
   - We calculate a matrix that shows how each numerical variable is related to every other variable. This helps us understand if there are patterns or connections between them.

3. **Rounding for Clarity:**
   - To make the numbers easier to read, we round the correlation matrix to two decimal places.

4. **Loading Necessary Package:**
   - We check if we have a tool called "reshape2" for data manipulation. If not, we install it.

5. **Data Preparation:**
   - We organize the correlation matrix into a format that's easy to use with a plotting tool called "ggplot2."

6. **Creating a Heatmap:**
   - We use a heatmap (a color-coded grid) to visually represent the correlation matrix. Darker colors show stronger correlations.

7. **Adjusting Heatmap Appearance:**
   - We make adjustments for a cleaner appearance, such as adding white borders to tiles, changing the color palette, and removing unnecessary elements like grid lines and ticks.

8. **Enhancements:**
   - We add a red dashed line to highlight the diagonal, which represents perfect correlation (a variable correlated with itself).

In summary, this code helps us visually explore how different numerical variables in the dataset are correlated with each other, providing insights into potential relationships.

```r
#### corrleation analysis
# Select only numerical columns for correlation analysis
numerical_columns <- sapply(cleaned_data, is.numeric)
correlation_data <- cleaned_data[, numerical_columns]

# Compute correlation matrix
correlation_matrix <- cor(correlation_data, use = "complete.obs")

# Round the correlation matrix to 2 digits
correlation_matrix <- round(correlation_matrix, digits = 2)

# Load the reshape2 package
if (!requireNamespace("reshape2", quietly = TRUE)) {
  install.packages("reshape2")
}
library(reshape2)

# Prepare data for ggplot2
melted_cormat <- melt(correlation_matrix)

# Plot the correlation matrix as a heatmap
library(ggplot2)

# Plot the correlation matrix as a heatmap with adjustments
ggplot(data = melted_cormat, aes(x = Var1, y = Var2, fill = value)) +
  geom_tile(color = "white") +  # Add white borders to tiles for better separation
  scale_fill_gradient(low = "lightblue", high = "darkblue") +  # Change color palette
  labs(title = "Correlation Matrix for Your Dataset") +
  theme_minimal() +
  theme(
    axis.text.x = element_text(angle = 90, hjust = 1, size = 8, color = "black"),  # Adjust text size and color
    axis.text.y = element_text(size = 8, color = "black"),  # Adjust text size and color
    plot.title = element_text(hjust = 0.5),  # Center the plot title
    panel.grid = element_blank(),  # Remove grid lines
    panel.border = element_blank(),  # Remove panel borders
    axis.ticks = element_blank(),  # Remove axis ticks
    legend.position = "right"  # Move legend to the right
  ) +
  geom_abline(slope = 1, intercept = 0, color = "red", linetype = "dashed", size = 0.5)  # Add diagonal line




```
